```python
'''
三维图
'''

def f(x, y):
    return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
xx, yy = np.meshgrid(x, y)
z = f(xx, yy)

# 普通3d图
ax = Axes3D(plt.figure())
ax.plot_surface(xx, yy, z)

# 等高线2d图
plt.figure()
plt.contourf(xx, yy, z, 20, alpha=.75) # 20等级, 绘制填充
cts = plt.contour(xx, yy, z, 20, colors='black') # 绘制轮廓线
plt.clabel(cts, inline=True, fontsize=10) # 添加高度标识
```

```python
'''
曲线拟合
'''
import scipy.optimize as opt

def fitfunc(x, a, b, c):
    return a*x**b+c

a, b, c = 1, 2, 3
xs = np.random.rand(20, 1)*5
ys = a*xs**b+c+(np.random.rand(20, 1)-0.5)*2
prm, cov = opt.curve_fit(fitfunc, xs, ys)

# plot
plt.scatter(xs, ys)
xf = np.linspace(0, 5, )
yt = fitfunc(xf, a, b, c)
yf = fitfunc(xf, prm[0], prm[1], prm[2])
plt.plot(xf, yt, '--')
plt.plot(xf, yf, 'r')
```
